```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./checkpoint/results.csv')
```

```python
print( df.columns)
df['error'] = 1 - df['best_acc'] 
df['lmacs'] = np.log(df['macs'])
```

```python
df.plot(x ='error', y='macs', kind = 'scatter')
```

```python
groups = df.groupby('model_name')

# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.error, group.macs, marker='o', linestyle='', ms=12, label=name)
ax.legend()
plt.yscale('log')
plt.title('CNN Accuracy vs MACS')
plt.show()
```
